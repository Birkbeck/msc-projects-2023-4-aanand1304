```python
# exploratory data analysis

import pandas as pd

# Load the dataset
file_path = 'FinalDataset.csv'
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Dropping the 'Date' column to focus on numerical features
df_numeric = df.drop(columns=['Date'])

# Calculate summary statistics for numerical features
summary_stats = df_numeric.describe().T

# Identify columns with missing values
missing_values = df_numeric.isnull().sum()

# Add missing values to the summary statistics
summary_stats['missing_values'] = missing_values

# Display the summary statistics
print(summary_stats)
```

```python
"""
Distribution Analysis
From this analysis, we can identify the distribution shape (e.g., normal, skewed), detect any outliers, 
and understand the spread and central tendency of each numerical feature in the dataset. 
This helps in assessing the suitability of the data for machine learning models and deciding on any necessary preprocessing steps.
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'FinalDataset.csv'
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Dropping the 'Date' column to focus on numerical features
df_numeric = df.drop(columns=['Date'])

# Set up the plotting environment
sns.set(style="whitegrid")

# Create a directory of plots for each numerical feature
for column in df_numeric.columns:
    plt.figure(figsize=(14, 6))

    # Histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df_numeric[column], kde=False, color='blue', bins=30)
    plt.title(f'Histogram of {column}')

    # Box Plot
    plt.subplot(1, 3, 2)
    sns.boxplot(y=df_numeric[column], color='green')
    plt.title(f'Box Plot of {column}')

    # Density Plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(df_numeric[column], color='red', shade=True)
    plt.title(f'Density Plot of {column}')

    # Adjust the layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python

"""
The correlation matrix shows whether variables move together (positive or negative correlation) 
and scatter plots visualize these relationships. 
This information is useful for selecting important features and understanding potential multicollinearity before building 
machine learning models.
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'FinalDataset.csv'
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Dropping the 'Date' column to focus on numerical features
df_numeric = df.drop(columns=['Date'])

# Calculate the correlation matrix
corr_matrix = df_numeric.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Matrix')
plt.show()

# Scatter plot for pairs of highly correlated variables
# Select pairs with high correlation (e.g., above 0.7)
high_corr_pairs = corr_matrix.unstack().sort_values(ascending=False).drop_duplicates()
high_corr_pairs = high_corr_pairs[(high_corr_pairs > 0.7) & (high_corr_pairs < 1)]

# Scatter plots for these pairs
for (var1, var2) in high_corr_pairs.index:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df_numeric[var1], y=df_numeric[var2])
    plt.title(f'Scatter Plot between {var1} and {var2}')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()
```

```python
#Missing Value Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'FinalDataset.csv'
df = pd.read_csv(file_path)

# Quantify missing values
missing_values = df.isnull().sum()
missing_values_percent = (missing_values / len(df)) * 100

# Create a dataframe to display missing values and their percentages
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_values_percent})
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)

# Visualize the extent of missing data
plt.figure(figsize=(12, 8))
sns.barplot(x=missing_data.index, y='Missing Values', data=missing_data, palette='viridis')
plt.xticks(rotation=90)
plt.title('Missing Values Count per Column')
plt.ylabel('Missing Values Count')
plt.show()

# Heatmap to visualize missing data
plt.figure(figsize=(16, 10))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap of Missing Values')
plt.show()
```

```python
#Outlier Detection

import pandas as pd
import numpy as np
from scipy.stats import zscore
# Load the dataset
file_path = 'FinalDataset.csv'
df = pd.read_csv(file_path)

# Dropping the 'Date' column to focus on numerical features
df_numeric = df.drop(columns=['Date'], errors='ignore')


z_scores = np.abs(zscore(df_numeric))
outliers_zscore = (z_scores > 3)

# Identifying the outliers
outlier_data_zscore = df_numeric[(outliers_zscore).any(axis=1)]

print("Outliers detected using Z-score:")
print(outlier_data_zscore)
```

```python
# IQR calculation
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))

# Identifying the outliers
outlier_data_iqr = df_numeric[(outliers_iqr).any(axis=1)]

print("Outliers detected using IQR:")
print(outlier_data_iqr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot for each numerical feature
plt.figure(figsize=(16, 10))
sns.boxplot(data=df_numeric)
plt.title('Box Plot for Outlier Detection')
plt.xticks(rotation=90)
plt.show()

# Pair plot to visualize outliers in pairwise relationships
sns.pairplot(df_numeric)
plt.title('Pair Plot for Outlier Detection')
plt.show()
```
