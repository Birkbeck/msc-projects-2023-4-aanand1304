```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Function to load and scale the data
def load_and_prepare_data(file_path, attack_column):
    data = pd.read_csv(file_path)
    attack_data = data[attack_column].values
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(attack_data.reshape(-1, 1))
    return scaled_data, scaler

# Function to convert time series to supervised learning format using a sliding window approach
def to_supervised(data, n_input, n_out=1):
    X, y = [], []
    for i in range(len(data) - n_input - n_out + 1):
        X.append(data[i:i + n_input])
        y.append(data[i + n_input:i + n_input + n_out])
    return np.array(X), np.array(y).squeeze()

# Function to build a complex model
def build_complex_model(n_input, n_features=1, complexity_level=1):
    model = Sequential()
    if complexity_level == 1:
        model.add(LSTM(50, activation='relu', input_shape=(n_input, n_features)))
    elif complexity_level == 2:
        model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features), return_sequences=True))
        model.add(Dropout(0.2))
        model.add(LSTM(50, activation='relu'))
    elif complexity_level == 3:
        model.add(Bidirectional(LSTM(100, activation='relu', return_sequences=True), input_shape=(n_input, n_features)))
        model.add(Dropout(0.2))
        model.add(Bidirectional(LSTM(50, activation='relu')))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mape'])
    return model

# Function to train the model and make predictions
def train_and_predict(data, n_input=12, n_out=1, epochs=100, batch_size=32, complexity_level=1):
    X, y = to_supervised(data, n_input, n_out)
    
    # Sliding window approach to create input-output pairs for training and testing
    split_position = len(data) - (36 + n_input)  # Last 3 years + lookback period
    X_train, X_test = X[:split_position], X[split_position:]
    y_train, y_test = y[:split_position], y[split_position:]

    model = build_complex_model(n_input, complexity_level=complexity_level)
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test), verbose=1)

    train_predict = model.predict(X_train)
    test_predict = model.predict(X_test)

    mae = mean_absolute_error(y_test, test_predict)
    mse = mean_squared_error(y_test, test_predict)
    rmse = np.sqrt(mse)

    return train_predict, test_predict, y_train, y_test, history, mae, mse, rmse

# Main execution
if __name__ == "__main__":
    file_path = 'FinalDataset.csv'
    attack_column = 'DDoS-ALL'  # Change this to the desired attack type

    scaled_data, scaler = load_and_prepare_data(file_path, attack_column)
    n_input = 12  # Lookback period
    
    for level in range(1, 4):  # Testing different complexity levels
        print(f"Training with complexity level {level}")
        train_predict, test_predict, y_train, y_test, history, mae, mse, rmse = train_and_predict(scaled_data, n_input, complexity_level=level)

        print(f"Level {level} - MAE: {mae:.3f}, MSE: {mse:.3f}, RMSE: {rmse:.3f}")

        plt.figure(figsize=(12, 6))
        plt.plot(scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual')
        plt.plot(scaler.inverse_transform(test_predict.reshape(-1, 1)), label='Predicted')
        plt.title(f'Level {level} - Predictions vs Actual')
        plt.legend()
        plt.show()
```
